```python
"""
Refactored Graph Algorithms module
- Self-contained (no external pip installs required)
- Implements: friend suggestions, Bellman-Ford, Dijkstra, Prim's MST
- Simple profiler using time + tracemalloc (for peak memory)
- Small demo dataset and plotting (matplotlib)

Save as graph_algorithms_refactor.py and run with: python graph_algorithms_refactor.py
"""

import time
import tracemalloc
import heapq
from collections import defaultdict, deque
import matplotlib.pyplot as plt


# ---------------------- Social / Utility Functions ----------------------

def suggest_friends(graph, user, max_suggestions=5):
    """Suggest friends for `user` based on number of mutual friends.
    Returns a list sorted by descending mutual-friend count.
    graph: dict[node] -> list[node]
    """
    if user not in graph:
        return []

    direct = set(graph[user])
    counts = defaultdict(int)

    # Count mutual friends: iterate user's direct friends and their friends
    for f in direct:
        for ff in graph.get(f, []):
            if ff != user and ff not in direct:
                counts[ff] += 1

    # sort by count (desc) then by name
    sorted_suggestions = sorted(counts.items(), key=lambda kv: (-kv[1], kv[0]))
    return [node for node, _ in sorted_suggestions[:max_suggestions]]


# ---------------------- Shortest Paths: Bellman-Ford ----------------------

def bellman_ford(edges, vertices, source):
    """Compute shortest paths from source on a graph given as edge list.
    edges: iterable of (u, v, w)
    vertices: iterable of vertex ids
    Returns dict vertex->distance or raises ValueError on negative cycle.
    """
    dist = {v: float('inf') for v in vertices}
    dist[source] = 0

    V = len(vertices)
    for _ in range(V - 1):
        updated = False
        for u, v, w in edges:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                updated = True
        if not updated:
            break

    # check for negative cycle
    for u, v, w in edges:
        if dist[u] + w < dist[v]:
            raise ValueError("Graph contains a negative-weight cycle")

    return dist


# ---------------------- Dijkstra (non-negative weights) ----------------------

def dijkstra(adj, source):
    """adj: dict[node] -> list of (neighbor, weight)
    Returns dict node->distance
    """
    dist = {n: float('inf') for n in adj}
    dist[source] = 0
    pq = [(0, source)]

    while pq:
        d, node = heapq.heappop(pq)
        if d > dist[node]:
            continue
        for nbr, w in adj.get(node, []):
            nd = d + w
            if nd < dist[nbr]:
                dist[nbr] = nd
                heapq.heappush(pq, (nd, nbr))

    return dist


# ---------------------- Prim's MST (using a start node) ----------------------

def prims_mst(adj, start):
    """adj: dict[node] -> list of (neighbor, weight)
    Returns total weight of MST covering reachable nodes from start.
    """
    visited = set()
    pq = [(0, start)]
    total = 0

    while pq:
        w, node = heapq.heappop(pq)
        if node in visited:
            continue
        visited.add(node)
        total += w
        for nbr, nw in adj.get(node, []):
            if nbr not in visited:
                heapq.heappush(pq, (nw, nbr))

    return total


# ---------------------- Simple profiler ----------------------

def profile(func, *args, **kwargs):
    """Return (result, elapsed_seconds, peak_memory_kb)
    peak memory measured by tracemalloc (peak - start) in KB.
    """
    tracemalloc.start()
    start_snapshot = tracemalloc.take_snapshot()
    t0 = time.perf_counter()
    result = func(*args, **kwargs)
    t1 = time.perf_counter()
    end_snapshot = tracemalloc.take_snapshot()
    tracemalloc.stop()

    # estimate peak as max difference across stats
    stats = end_snapshot.compare_to(start_snapshot, 'lineno')
    peak_bytes = sum(s.size_diff for s in stats if s.size_diff > 0)
    peak_kb = peak_bytes / 1024
    return result, (t1 - t0), peak_kb


# ---------------------- Demo datasets & plotting ----------------------

def demo_and_plot():
    # Social graph example
    social = {
        'A': ['B', 'C'],
        'B': ['A', 'D'],
        'C': ['A', 'D', 'E'],
        'D': ['B', 'C'],
        'E': ['C']
    }

    # Route graph as edges for Bellman-Ford
    vertices = ['A', 'B', 'C', 'D']
    edges = [
        ('A', 'B', 4),
        ('A', 'C', 2),
        ('C', 'B', -1),
        ('B', 'D', 2),
        ('C', 'D', 3),
    ]

    # Graph for Dijkstra / Prim (adjacency)
    adj = {
        'A': [('B', 1), ('C', 4)],
        'B': [('C', 2), ('D', 5)],
        'C': [('D', 1)],
        'D': []
    }

    # Prim adjacency (undirected)
    prim_adj = {
        'A': [('B', 2), ('C', 3)],
        'B': [('A', 2), ('C', 1), ('D', 4)],
        'C': [('A', 3), ('B', 1), ('D', 5)],
        'D': [('B', 4), ('C', 5)]
    }

    print("=== DEMO OUTPUT ===")
    print("Suggested friends for A:", suggest_friends(social, 'A'))

    bf_res, bf_time, bf_mem = profile(bellman_ford, edges, vertices, 'A')
    print("Bellman-Ford distances:", bf_res)

    d_res, d_time, d_mem = profile(dijkstra, adj, 'A')
    print("Dijkstra distances:", d_res)

    p_res, p_time, p_mem = profile(prims_mst, prim_adj, 'A')
    print("Prim's MST total cost:", p_res)

    # plotting times and memory (simple bar chart)
    names = ['Bellman-Ford', 'Dijkstra', "Prim's MST"]
    times = [bf_time, d_time, p_time]
    mems = [bf_mem, d_mem, p_mem]

    plt.figure(figsize=(7, 4))
    plt.bar(names, times)
    plt.title('Execution time (seconds)')
    plt.ylabel('Seconds')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(7, 4))
    plt.bar(names, mems)
    plt.title('Estimated memory delta (KB)')
    plt.ylabel('KB')
    plt.tight_layout()
    plt.show()


# ---------------------- Summary Table ----------------------

def summary_table():
    rows = [
        ("Social Network Suggestion", "Mutual friends", "O(V + E)", "Social Media"),
        ("Route Finding", "Bellman-Ford", "O(VE)", "Navigation"),
        ("Emergency Response", "Dijkstra", "O(E log V)", "Disaster Mgmt"),
        ("Cable Installation", "Prim's", "O(E log V)", "Infrastructure"),
    ]

    print('\nFINAL SUMMARY')
    print('-' * 60)
    print(f"{'Problem':<28}{'Algorithm':<18}{'Time Complexity':<12}{'Domain'}")
    print('-' * 60)
    for p, a, t, d in rows:
        print(f"{p:<28}{a:<18}{t:<12}{d}")
    print('-' * 60)


# ---------------------- Main ----------------------


def main():
    demo_and_plot()
    summary_table()


if __name__ == '__main__':
    main()
```
